### Imports

```python
import pandas as pd
import numpy as np
import os
import shutil
```

```python
df=pd.read_csv("LibriSpeechTrain/SPEAKERS_MODIFIED.txt",sep="|")
```

```python
df.info()
```

```python
df.columns
```

```python
#Stripping spaces from the columns name
df.columns=[i.strip() for i in df.columns]
df.columns
```

```python
#Striping spaces from the data values except columns that caontain int and float values
for i in df.columns[1:3]:
    df[i]=df[i].apply(lambda x:x.strip())
for i in df.columns[4:]:
    df[i]=df[i].apply(lambda x:x.strip())
```

```python
df['SUBSET'].value_counts().axes[0][2]
```

```python
df=df[df['SUBSET']=="train-clean-100"]
```

```python
df.shape
```

```python
df.head()
```

```python
df.drop(["SUBSET","MINUTES","NAME"],inplace=True,axis=1)
```

```python
df.reset_index(drop=True)
```

```python
file_path="LibriSpeechTrain/train-clean-100"
```

```python
#Making subdirectories
os.mkdir("male")
os.mkdir("female")
```

```python
def make_folders(files):
    
    z=0
    if files.SEX=="M":
        z=1
    print(files.ID,files.SEX,z)
    
    # Sets the name to be the path to where the file is in my computer
    file_name = os.path.join(os.path.abspath("LibriSpeechTrain/train-clean-100"+'/'+str(files.ID)))
    dir_list=os.listdir(file_name)
    if z==1:
        for i in dir_list:
            for j in os.listdir(file_name+"/"+i):
                #shutil.copy(src,dest)
                shutil.copy(file_name+"/"+i+"/"+j,"male/")
    else:
        for i in dir_list:
            for j in os.listdir(file_name+"/"+i):
                #shutil.copy(src,dest)
                shutil.copy(file_name+"/"+i+"/"+j,"female/")
    
                             
```

```python
df.apply(make_folders,axis=1)
```

```python
#Removinf unnecessary text files from male and female directory
for i in os.listdir('male/'):
    if i[-3:]=='txt':
        os.remove("male/"+i)
        print(i)
for i in os.listdir('female/'):
    if i[-3:]=='txt':
        os.remove("female/"+i)
        print(i)
```

```python
#Directory that contain all the train audio files
os.mkdir("voice")
```

```python
#copying all files in a single directory named "voice"
for i in os.listdir("male"):
    shutil.copy("male/"+i,"voice/")
```

```python
for i in os.listdir("female"):
    shutil.copy("female/"+i,"voice/")
```
